```python
#grouping
gdeptid=df.groupby('DEPARTMENT_ID')
```

```python
#find sum of salary department wise
sumsal=gdeptid['SALARY'].sum()
print(sumsal)
```

```python
type(sumsal)
```

```python
#print the department whose total salary > 10000
print(sumsal[sumsal.loc[:]>10000].count())
```

```python
#display the avg salary
meansal=gdeptid['SALARY'].mean()
print(meansal)
```
